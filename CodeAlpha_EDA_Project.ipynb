### Objective
The objective of this project is to perform Exploratory Data Analysis (EDA) on a global layoffs dataset to identify trends, patterns, and business insights using Python. The dataset contains information about layoffs in various industries and countries.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("layoffs.csv")
df.head()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Q1. which company has the highest number of layoffs?

top_company = df.groupby('company')['total_laid_off'].sum().sort_values(ascending=False).head(10)

print(top_company)
```

```python
# visualization
top_company.plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Companies with  Layoffs')
plt.xlabel('Company')
plt.ylabel('Employees Laid Off')

plt.show()
```

```python
## Q2 which industry  affected most?

industry = df.groupby('industry')['total_laid_off'].sum().sort_values(ascending=False).head(10)
industry.plot(kind='bar',figsize=(10,5))

plt.title("Most Affected Industries")

plt.show()

```

```python
## which country had most layoffs?

country = df.groupby('country')['total_laid_off'].sum().sort_values(ascending=False).head(10)
country.plot(kind='pie')
plt.title(" Layoffs by Country")
plt.ylabel("")
plt.show()
```

```python
# time series analysis
df['date'] = pd.to_datetime(df['date'])
```

```python
monthly = df.groupby(df['date'].dt.to_period('M'))['total_laid_off'].sum()

monthly.plot(figsize=(12,5))

plt.title("Monthly Layoff Trend")

plt.xlabel("Month")

plt.ylabel("Layoffs")

plt.show()
```

# Key Insights

- Tech companies experienced the highest layoffs.
- United States recorded maximum layoffs globally.
- Layoffs increased significantly during recession periods.
- Several funded startups also reduced workforce.
